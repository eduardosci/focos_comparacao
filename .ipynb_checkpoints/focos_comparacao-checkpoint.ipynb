* Nesse notebook analiso a comparação de focos de incendios ocorridos entre os periodos de janeiro a agosto dos anos 2020 x 2021

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
import folium
from folium import plugins
import pydeck as pdk
from mpl_toolkits.basemap import Basemap
import plotly.offline as py
import cufflinks as cf
cf.go_offline()
import sweetviz as sv
```

```python
data = pd.read_csv('/home/eduardo/projetos/analise_queimadas/fire_2020.csv')
```

```python
data2 = pd.read_csv('/home/eduardo/projetos/analise_queimadas/fire_2021.csv')
```

```python
data['acq_date'] = pd.to_datetime(data['acq_date'],errors='ignore') 
```

```python
data2['acq_date'] = pd.to_datetime(data2['acq_date'],errors='ignore') 
```

```python
sel1 = (data['acq_date'] >= '2020-01-01') & (data['acq_date'] <= '2020-08-31')
```

```python
df_fire_2020 = data[sel1]
```

```python
df_fire_2020.dtypes
```

```python
sel2 = (data2['acq_date'] >= '2021-01-01') & (data2['acq_date'] <= '2021-08-31')
```

```python
df_fire_2021 = data2[sel2]
```

```python
df_fire_2021.dtypes
```

```python
min(df_fire_2020['acq_date']), max(df_fire_2020['acq_date'])
```

```python
min(df_fire_2021['acq_date']), max(df_fire_2021['acq_date'])
```

```python
df_fire_2020['fire_month'] = df_fire_2020.acq_date.dt.to_period('M').astype(str)
```

```python
df_fire_2021['fire_month'] = df_fire_2021.acq_date.dt.to_period('M').astype(str)
```

```python
focos_mes_2020 = df_fire_2020.groupby(by='fire_month').brightness.sum()
```

```python
focos_mes_2021 = df_fire_2021.groupby(by='fire_month').brightness.sum()
```

```python
data = [go.Bar(x=focos_mes_2020.index,
               y=focos_mes_2020.values,
               marker = {'color': "orange",
                         'line': {'color': '#333',
                                  'width': 2}
                        },
               opacity= 0.7,
               name='2020'
              ),
# definindo outro plot de barras com valores dos meses passados
        
        go.Bar(x=focos_mes_2021.index,
               y=focos_mes_2021.values,
               name='2021',
               marker = {'color': 'lightgreen',
                         'line': {'color': '#333',
                                  'width': 2}
                        },
               opacity= 0.7
              )
              
       ]

# Criando Layout
configuracoes_layout = go.Layout(title='Focos de incendios entre os periodos de janeiro a agosto de 2020 e janeiro a agosto de 2021',
                                 yaxis={'title':'Quantidade De Incendios'},
                                 xaxis={'title': 'Periodo'}                    
                                
                                )

# Objeto figura

fig = go.Figure(data=data, layout=configuracoes_layout)

# plotando o grafico
py.iplot(fig, filename='Focos no periodo 2020 e 2021')
```

```python
qtd_jan_ago_2020 = df_fire_2020.query("acq_date >= '2020-01-01' and acq_date <= '2020-08-31'").shape[0]
qtd_jan_ago_2021 = df_fire_2021.query("acq_date >= '2021-01-01' and acq_date <= '2021-08-31'").shape[0]
aumento_qtd_jan_ago = qtd_jan_ago_2021 - qtd_jan_ago_2020
aumento_porcent_jan_ago = aumento_qtd_jan_ago / qtd_jan_ago_2020 * 100
```

```python
qtd_jan_ago_2020, qtd_jan_ago_2021
```

```python
aumento_qtd_jan_ago, '{:.2f}%'.format(aumento_porcent_jan_ago)
```

#### Foi constatado um aumento de 3.8% no período entre os anos 2020x2021, uma diferença de -2% em relação ao descrito na matéria. Com isso, fica confirmado um aumento entre os períodos de cada ano, porém com uma taxa menor do que o descrito na reportagem. Há a possibilidade da fonte de dados utilizada na matéria não ser a mesma deste estudo e isto causar a diferença.

```python
qtd_ago_2020 = df_fire_2020.query("acq_date >= '2020-08-01' and acq_date <= '2020-08-31'").shape[0]
qtd_ago_2021 = df_fire_2021.query("acq_date >= '2021-08-01' and acq_date <= '2021-08-31'").shape[0]
aumento_qtd = qtd_ago_2021 - qtd_ago_2020
aumento_porcent = aumento_qtd / qtd_ago_2020 * 100
```

```python
aumento_qtd, '{:.2f}%'.format(aumento_porcent)
```

#### Foi constatado um aumento de 6% no período  do mes de agosto de 2021 comparado ao mes de agosto de 2021, uma diferença de -11% em relação ao descrito na matéria. Isso confirma um aumento entre os períodos , porém tambem com uma taxa  bem menor do que o descrito na reportagem. Sabendo que existe a possibilidade da fonte de dados utilizada na matéria não ser a mesma deste estudo, o que justifica essa diferença.

```python
plot = sns.barplot(x=[qtd_jan_ago_2020, qtd_jan_ago_2021], y=[qtd_jan_ago_2020, qtd_jan_ago_2021])
for i in plot.patches:
    plot.annotate(i.get_height(),
                 (i.get_x() + i.get_width() / 2, i.get_height()),
                 ha='center', va='baseline', fontsize=10, color='black',
                 xytext=(0,2), textcoords='offset pixels')
    plt.title('Focos de incendios entre janeiro a agosto dos anos 2020 x 2021', fontsize=15)
    plt.xlabel('Quantidade', fontsize=15)
    plt.text(-0.1, 500000, '2020', fontsize = 15)
    plt.text(0.9, 500000, '2021', fontsize = 15)
```

#### Nesse grafico pode-se observar mais claramente a diferença da quantidade dos focos de incendios de cada ano.
